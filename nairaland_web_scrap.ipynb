```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import seaborn as sns
import datetime
```

```python
date_today = str(datetime.date.today())
```

```python
html_response = requests.get('https://nairaland.com/')
```

```python
html = html_response.text
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
soup.title.text
```

```python
soup.div.prettify()
```

```python
users_birthday = soup.find('td', class_='homeuser')
```

```python
users_birthday = users_birthday.text
```

```python
users_birthday = users_birthday.split(':')
```

```python
users_birthday = users_birthday[1:]
```

```python
bd_list = users_birthday[0].split(',')
```

```python
user_data = {}
csv_file = open('nairaland_users_birthday.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['username','age','date'])


date = date_today

for i in range(len(bd_list)):
    split_list = bd_list[i].split('(')
    name = split_list[0]
    age = ''
    try:
        age = split_list[1]
        split_list = split_list[1].split(')')
        age = split_list[0]
        print(name,age)
        
    except:
        print(name)
    user_data[name] = age
    csv_writer.writerow([name,age,date])
csv_file.close()
```

```python
df = pd.read_csv('nairaland_users_birthday.csv')
```
